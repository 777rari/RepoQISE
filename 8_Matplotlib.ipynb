We will first try and plot the sine function using matplotlib

```python
# our sine function
def sin(x:float)-> float:
    return np.sin(x)
```

```python
# plot the sine function over the time t
t = np.arange(0,10,step=0.01)

# YOUR CODE STARTS HERE

plt.plot(t, sin(t),label="your line name", color="red", linestyle="dashed" )
plt.title("____")
plt.xlabel("____")
plt.ylabel("____")
plt.legend() # shows your label
plt.show() # makes sure the plot is showing

# YOUR CODE ENDS HERE
```
